```python
import json
import math
import numpy as np
from scipy.optimize import minimize
from scipy.optimize import minimize_scalar
import matplotlib.pyplot as plt

# zmienne globalne (jednostki SI):

# parametry kuli
V0 = 10                            # predkosc kuli

# zmiene do sily oporu L 
ro = 1.1225     # gestosc powietrza
Cd = 0.47       # wspolczynnik sily oporu
g = 9.81        # przyspieszenie grawitacyjne

# odczytanie R i h0
def read_data():
    data = []
    data_big = []  # Lista, do której będziemy dodawać odczytane wartości

    # Otwórz plik do odczytu
    with open('/root/DANE.txt', 'r') as file:
    
        for linia in file:
            linia = linia.strip()  # Usunięcie białych znaków z początku i końca linii
            if linia:
                d_1 = linia.split(',')
                data = [d_2.strip() for d_2 in d_1]
                data[0] = float(data[0])
                data[1] = float(data[1])
                data_big.append(data)

    return(data_big)

# funkcja obliczająca trajektorie z równań ruchu uwzględniająca zmieniającą się prędkość
def Trajectory (t, V0, phi, A, Cd, ro, g, m, h0, Vy0, Vx0, t_prev):
    Vy = V0 * math.sin(phi) - g * t - Vy0 **2 * ro * A * Cd * t / m
    Vx = V0 * math.cos(phi) - Vx0 **2 * ro * A * Cd * t / m
    h = h0 + Vy0 * t_prev - 0.5 * g * t **2 - 0.5 * (Vy0 **2) * ro * A * Cd * (t**2) / m     
    d = Vx0 * (t) - 0.5 * Vx0 **2 * ro * A * Cd * (t **2) / m
    return Vy, Vx, h, d

# funkcja zwracająca zasięg
def Find_d_max(phi, R, h0):
    m = 4/3 * math.pi * (R**3) * 7850   # masa kuli
    A = math.pi*R **2                   # przekrój kuli
    t_prev = 0
    Vy_prev = V0 * math.sin(phi)
    Vx_prev = V0 * math.cos(phi)
    time = np.linspace(0, 10, 1000)
    for t in time:
        Vy, Vx, Y, X = Trajectory(t, V0, phi, A, Cd, ro, g, m, h0, Vy_prev, Vx_prev, t_prev)
        t_prev = t
        Vy_prev = Vy
        Vx_prev = Vx
        if Y < 0.01  :
            return X

# funkcja do rysowania trajektorii
def Draw_trajectory(phi, R, h0):
    m = 4/3 * math.pi * (R**3) * 7850   # masa kuli
    A = math.pi*R **2                   # przekrój kuli
    t_prev = 0
    Vy_prev = V0 * math.sin(phi)
    Vx_prev = V0 * math.cos(phi)
    Yks = []
    times = []
    Vyks = []
    time = np.linspace(0, 10, 1000)
    for t in time:
        Vy, Vx, Y, X = Trajectory(t, V0, phi, A, Cd, ro, g, m, h0, Vy_prev, Vx_prev, t_prev)
        t_prev = t
        Vy_prev = Vy
        Vx_prev = Vx
        Yks.append(Y)
        times.append(t)
        Vyks.append(Vy_prev)
        if Y < 0.01  :
            return  Yks, times

results = []
data_big = read_data()

for i in range (0,13):
    data = data_big[i*5]
    print("Dla promienia R=" , data[0])
    fig, axs = plt.subplots(1, 2, figsize=(30, 9))  # Tworzenie jednego wiersza i pięciu kolumn subplotów
    
    for j in range(i*5, i*5+5): # dla każdej wartości promienia:
        data = data_big[j]
        result = minimize_scalar(lambda phi: -Find_d_max (phi,data[0], data[1]), bounds=(0, math.pi), method='bounded')             #maksymalizacja po kącie
        print("Dla wysokości wyrzutu h0=" , data[1] , "m: Optymalny kąt:" , round(math.degrees(result.x),2),"Zasięg:",-result.fun)
        
        result_dict = {
            "promien_R": data[0],
            "wysokosc_h0": data[1],
            "optymalny_kat": round(math.degrees(result.x), 2),
            "zasieg": -result.fun
        }

        results.append(result_dict)
        
        angle = np.linspace(0, math.pi/2, 900)
        ds = []
        trajectories = []
        times = []

        for a in angle:
            ds.append(Find_d_max (a,data[0], data[1]))
        trajectories, times = Draw_trajectory(result.x,data[0], data[1])
        axs[0].plot(np.degrees(angle), ds, label=f"Promień R={data[0]}, wysokość h0={data[1]}")  # Rysowanie wykresu
        axs[1].plot(times, trajectories, label=f"Promień R={data[0]}, wysokość h0={data[1]}")    # Rysowanie wykresu

    
    # rysowaine wykresów zasięgu od kąta i trajektorii dla danego promienia kuli na dwóch oddzielych wykresach
    axs[0].set_xlabel("Kąt (stopnie)")  # Etykieta osi x
    axs[0].set_ylabel("Zasięg")  # Etykieta osi y
    axs[0].legend()  # Dodanie legendy
    axs[1].set_xlabel("Czas (s)")  # Etykieta osi x
    axs[1].set_ylabel("Trajektoria")  # Etykieta osi y
    axs[1].legend()  # Dodanie legendy
    plt.show()  # Wyświetlenie wykresu       

# serializacja
with open("wyniki.json", "w") as f:
    json.dump(results, f)
```
